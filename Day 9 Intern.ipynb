```python
# TASKS (Using Movie Success Analysis(Task 1: Budget vs Rating))

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv(r"C:\Users\Alana\Downloads\movies.csv.zip")

# Display First 5 Records
print(df.head())

# Check Missing Values
print(df[['budget','score']].isnull().sum())

# Remove Missing Values
df = df.dropna(subset=['budget','score'])

# Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='budget', y='score')

plt.title("Budget vs Movie Rating")
plt.xlabel("Budget")
plt.ylabel("IMDb Score")

plt.show()
```

```python
# TASKS (Using Movie Success Analysis (Task 2: Genre Popularity))

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv(r"C:\Users\Alana\Downloads\movies.csv.zip")

# Count Genres
genre_count = df['genre'].value_counts()

print(genre_count)

# Plot
plt.figure(figsize=(10,6))

genre_count.plot(kind='bar')

plt.title("Genre Popularity")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")

plt.show()
```

```python
# TASKS (Using Movie Success Analysis (Task 3: Director Performance))

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv(r"C:\Users\Alana\Downloads\movies.csv.zip")

# Average Rating of Directors
director = df.groupby('director')['score'].mean()

# Top 10 Directors
top_director = director.sort_values(ascending=False).head(10)

print(top_director)

# Plot
plt.figure(figsize=(10,6))

top_director.plot(kind='bar')

plt.title("Top 10 Directors by Average Rating")
plt.xlabel("Director")
plt.ylabel("Average IMDb Score")

plt.show()
```

```python
# TASKS (Using Movie Success Analysis(Task 4: Revenue Trends))

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv(r"C:\Users\Alana\Downloads\movies.csv.zip")

# Group Revenue by Year
revenue = df.groupby('year')['gross'].sum()

print(revenue)

# Plot
plt.figure(figsize=(10,6))

revenue.plot()

plt.title("Revenue Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Total Revenue")

plt.grid(True)

plt.show()
```

```python
# TASKS (Using Movie Success Analysis(Task 5: Movie Recommendation Insights))

# Import Libraries
import pandas as pd

# Load Dataset
df = pd.read_csv(r"C:\Users\Alana\Downloads\movies.csv.zip")

# Remove Missing Values
df = df.dropna(subset=['score','gross','budget'])

# Recommend Movies
recommend = df[
    (df['score'] >= 8.0) &
    (df['gross'] >= 100000000) &
    (df['budget'] >= 50000000)
]

print("Recommended Movies")

print(recommend[['name',
                 'genre',
                 'director',
                 'score',
                 'budget',
                 'gross']])
```
